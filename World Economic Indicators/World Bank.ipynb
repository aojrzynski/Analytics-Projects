# World Economic Report
This project aims to provide a report on the state of the global economy by analyzing datasets from the Human Development Index (HDI) and the World Bank. The goal is to highlight key insights and trends that are shaping the economic landscape across different regions and countries.

## About the datasets
**Human Development Index (HDI) Dataset**

This dataset provides HDI rankings and scores for countries from 1990 to 2021. It includes various indicators that contribute to the HDI, such as life expectancy, education, and per capita income indices.

**World Bank Dataset**

This dataset includes multiple sheets with economic indicators collected by the World Bank for various countries and years. Key indicators include GDP, internet usage, infant mortality rates, life expectancy, population density, and unemployment rates.

## Data Dictionary

### Human Development Index Dataset

| Column        | Description                                                      |
|---------------|------------------------------------------------------------------|
| `iso3`        | ISO 3-letter country code.                                       |
| `country`     | Name of the country.                                             |
| `hdicode`     | Classification based on HDI (e.g., Low, Medium, High, Very High).|
| `region`      | The region to which the country belongs.                         |
| `hdi_rank_2021`| HDI rank of the country in 2021.                                |
| `hdi_<year>`  | HDI score for the given year.                                    |
| `mf_<year>`   | Various other indicators for the given year.                     |

### World Bank Dataset

| Column                                                      | Description                                                                     |
|-------------------------------------------------------------|---------------------------------------------------------------------------------|
| `Country Name`                                              | The name of the country.                                                        |
| `Country Code`                                              | The three-letter code representing the country.                                 |
| `Region`                                                    | The World Bank region of the country.                                           |
| `IncomeGroup`                                               | The World Bank income group classification of the country.                      |
| `Year`                                                      | The year in which the statistics were recorded.                                 |
| `GDP (current US$)`                                         | The Gross Domestic Product in current US dollars.                               |
| `GDP per capita (USD)`                                      | GDP per capita in US dollars.                                                   |
| `Individuals using the Internet (% of population)`          | Percentage of the population using the internet.                                |
| `Infant mortality rate (per 1,000 live births)`             | The number of infant deaths per 1,000 live births.                              |
| `Life expectancy at birth (years)`                          | Average life expectancy at birth.                                               |
| `Population density (people per sq. km of land area)`       | Population density measured in people per square kilometer of land area.        |
| `Unemployment (% of total labor force)`                     | Unemployment rate as a percentage of the total labor force.                     |



# Prepare & QA the Data

```python
import pandas as pd
import numpy as np

# Load the World Bank dataset and rename 'Country Name' to 'Country'
world_bank = pd.read_excel("WorldBank.xlsx").rename({"Country Name": "Country"})

# Calculate the population in millions and add it as a new column
world_bank["Population (M)"] = (world_bank["GDP (USD)"] / world_bank["GDP per capita (USD)"]) / 1_000_000

# Inspect the dataset
world_bank.head()
```

```python
# Check column types and non-null values
world_bank.info()
```

```python
# Display descriptive statistics for the dataset
world_bank.describe()
```

```python
# Load the Human Development Index (HDI) dataset
hdi = pd.read_csv("hdi.csv")

# Filter the World Bank dataset for the year 2014 and merge it with the HDI dataset
wb_hdi_2014 = world_bank.query("Year == 2014").merge(hdi[["iso3", "hdi_2014"]], how="left", left_on="Country Code", right_on="iso3")

# Inspect the dataset
wb_hdi_2014.head()
```

```python
# Display descriptive statistics for the dataset
wb_hdi_2014.describe()
```

# Prepare the data for visualization

```python
# Create a pivot table to sum GDP (USD) by Year and Region
gdp_pivot = world_bank.pivot_table(index="Year", columns="Region", values="GDP (USD)", aggfunc="sum")

# Inspect the dataset
gdp_pivot.head()
```

```python
# Create a pivot table to sum Population (M) by Year and Region
pop_pivot = world_bank.pivot_table(index="Year", columns="Region", values="Population (M)", aggfunc="sum")

# Inspect the dataset
pop_pivot.head()
```

```python
# Group the merged World Bank and HDI dataset by Region and calculate the mean HDI for 2014
# Sort by descending order
wb_hdi_by_region = wb_hdi_2014.groupby("Region").agg({"hdi_2014": "mean"}).sort_values("hdi_2014", ascending=False) 

# Inspect the dataset
wb_hdi_by_region.head()
```

# Visualize the data

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mticker
```

```python
# Create a stack plot for GDP over time by region
fig, ax = plt.subplots()

# Plot GDP in trillions for each region sorted by GDP value in the last year available
ax.stackplot(
    gdp_pivot.index,
    [gdp_pivot[region] / 1_000_000_000_000 for region in gdp_pivot.iloc[-1].sort_values(ascending=False).index],
    labels = gdp_pivot.iloc[-1].sort_values(ascending=False).index,
);

ax.legend(loc="upper left")
ax.set_title("GDP has Grown Exponentially Over Time")
ax.set_ylabel("GDP (Trillions)");
```

```python
# Create a stack plot for Population over time by region
fig, ax = plt.subplots()

# Plot Population in billions for each region sorted by population value in the last year available
ax.stackplot(
    pop_pivot.index,
    [pop_pivot[region] / 1000 for region in pop_pivot.iloc[-1].sort_values(ascending=False).index],
    labels = pop_pivot.iloc[-1].sort_values(ascending=False).index,
);

ax.set_title("Population has surged from 2 billion to 7.5 billion")
ax.set_ylabel("Population (Billions)");
```

```python
# Create a scatter plot to show the relationship between Life Expectancy and GDP per capita
fig, ax = plt.subplots()

minsize = min(wb_hdi_2014["Population (M)"])
maxsize = max(wb_hdi_2014["Population (M)"])

sns.scatterplot(
data=wb_hdi_2014,
x="Life expectancy at birth (years)",
y="GDP per capita (USD)",
size="Population (M)",
sizes=(minsize, maxsize),
hue="Region",
).set(title="Life Expectancy Increases as Countries get richer", yscale="log");

ax.yaxis.set_major_formatter(mticker.ScalarFormatter())
```

```python
# Plot HDI by region as a bar chart
wb_hdi_by_region.plot.bar(title="HDI by Region", ylabel="Human Development Index (HDI)")
```

```python
# Create a scatter plot to show the relationship between Electric Power Consumption and GDP per capita
sns.scatterplot(
    data=wb_hdi_2014.query("`Country Name` != 'Iceland'"),
    x="Electric power consumption (kWh per capita)",
    y="GDP per capita (USD)",
    hue="hdi_2014",
    palette="coolwarm_r"
).set(title="Electricity Drives Development")
```

# Build a final report

```python
import matplotlib.gridspec as gridspec
import matplotlib as mpl
```

```python
# Set up the figure and grid specification
fig = plt.figure(figsize=(8,12))
gs = gridspec.GridSpec(nrows=12, ncols=6)
mpl.rcParams['font.family'] = "Lato"

# Define color palettes for different plots
stack1_list = ['#66C2A5', '#fC8D62', '#8DA0CB', '#E78AC3', '#A6D854', '#FFD92F', '#B3B3B3']
stack2_list = ['#66C2A5', '#A6D854', '#B3B3B3', '#fC8D62', '#E78AC3', '#8DA0CB', '#FFD92F']
bar_list = ['#8DA0CB', '#fC8D62', '#E78AC3', '#FFD92F', '#66C2A5', '#A6D854', '#B3B3B3']
bubble_list = ['#A6D854', '#fC8D62', '#FFD92F', '#66C2A5', '#B3B3B3', '#E78AC3', '#8DA0CB']

# Set the main title of the report
fig.suptitle(
    "Report: Global Economic Development from 1960-2018",
    fontsize = 16,
    fontweight = "bold",
    x=.65,
    y=.93
)


# Create the GDP stack plot
ax1 =  fig.add_subplot(gs[0:4, 0:3])

ax1.stackplot(
    gdp_pivot.index,
    [gdp_pivot[region] / 1_000_000_000_000 for region in gdp_pivot.iloc[-1].sort_values(ascending=False).index],
    labels = gdp_pivot.iloc[-1].sort_values(ascending=False).index,
    colors=stack1_list
);

ax1.set_title("GDP has Grown Exponentially Over Time")
ax1.set_ylabel("GDP (Trillions)");


# Add explanatory text
fig.text(.92, .62,
'''
Economic expansion in the 20th and early
21st centuries has been enormous and has
outpaced population growth. In 1960, Europe 
and North America produced 75% of GDP with 
20% of the population.

By 2018, East Asia had surpassed both 
in economic output largely due to the
rise of Japan, South Korea and China.

In the coming decades, economists expect
South Asia and Sub-Saharan Africa to lead
the world in global growth.
'''
, fontsize=8)


# Create the Population stack plot
ax2 =  fig.add_subplot(gs[0:4, 3:6])

ax2.stackplot(
    pop_pivot.index,
    [pop_pivot[region] / 1000 for region in pop_pivot.iloc[-1].sort_values(ascending=False).index],
    labels = pop_pivot.iloc[-1].sort_values(ascending=False).index,
    colors=stack2_list
);

ax2.set_title("Population has surged from 2 billion to 7.5 billion")
ax2.set_ylabel("Population (Billions)");
ax2.legend(bbox_to_anchor=(1.75, 1), fontsize=8)


# Create the scatter plot for Life Expectancy vs GDP per capita
ax3 =  fig.add_subplot(gs[4:8, :])

minsize = min(wb_hdi_2014["Population (M)"])
maxsize = max(wb_hdi_2014["Population (M)"])

sns.scatterplot(
    data=wb_hdi_2014,
    x="Life expectancy at birth (years)",
    y="GDP per capita (USD)",
    size="Population (M)",
    sizes=(minsize, maxsize),
    hue="Region",
    palette=bubble_list,
    ax=ax3
).set(title="Life Expectancy Increases as Countries get richer", yscale="log");

ax3.yaxis.set_major_formatter(mticker.ScalarFormatter())

# Adjust legend for scatter plot
handles, labels = ax3.get_legend_handles_labels()
entries_to_skip = len(wb_hdi_2014["Region"].unique())+1

for h in handles[1:]:
    sizes = [s / 1.5 for s in h.get_sizes()]
    h.set_sizes(sizes)

ax3.legend(
    handles[entries_to_skip:], 
    labels[entries_to_skip:], 
    bbox_to_anchor=(.8, .9), 
    borderaxespad=0.,
    frameon=False,
    ncol=6,
    fontsize=8
)


# Add explanatory text
fig.text(.92, .45,
'''
The wealthy nations of the world
enjoy high GDP per capita as well 
as long life spans, but make up
a relatively small share of the
global population.

If growth can continue in the developing
world, humanity will be vastly wealthier,
healthier and (hopefully) happier.
'''
, fontsize=8)


# Create the bar plot for HDI by region
ax4 = fig.add_subplot(gs[8:12, 0:3])

wb_hdi_by_region.plot.bar(ax = ax4, title="HDI by Region", ylabel="Human Development Index (HDI)", color=bar_list)

# Create the scatter plot for Electricity Consumption vs GDP per capita
ax5 =  fig.add_subplot(gs[8:12, 3:6])

sns.scatterplot(
    data=wb_hdi_2014.query("`Country Name` != 'Iceland'"),
    x="Electric power consumption (kWh per capita)",
    y="GDP per capita (USD)",
    hue="hdi_2014",
    palette="coolwarm_r",
    ax=ax5
).set(title="Electricity Drives Development")


# Add explanatory text
fig.text(.92, .15,
'''
HDI, short for Human Development Index,
attempts to measure the overall standard
of living in countries. Life Expectancy,
GDP per Capita, and Educational attainment
are the factors that are considered.

Economic growth in the developing world
in the 21st century should help other regions
catch up to North America and Europe.

One factor not in HDI, but instrumental in
a country's development, is electricity consumption.
Electricity unlocks massive improvements in
productivity. Developing nations should continue
to invest in energy production to ensure growth.
'''
, fontsize=8)


# Adjust subplot spacing
fig.subplots_adjust(wspace=4, hspace=4)
```
